```python
# Functions In Python
# A function is a block of code which only runs when it is called.
# Functions provide better modularity for your application and a high degree of code reusing.
# Good thing about functions is that they are famous with several names. Different programming
# languages name them differently, for example, functions, methods, sub-routines, procedures, etc.
# Functions make your code cleaner and save you a lot of time.
# Python Functions stop us from writing the same logic various times.
# We can bind the logic in one function and then call the same repeatedly.
# In Python a function is defined using the def keyword

# Syntax

 # def function_name():
 #     code to be executed

# So, why should you learn to use functions?
# Here are 5 clear reasons:
# Avoid repetition (Don't Repeat Yourself – DRY)
# Make your code easier to read
# Easier to fix and debug
# Break complex problems into smaller parts
# Reuse code in different places or even in different programs
```

```python
# function without parameter/argument

#function definition

def greetuser():
  print("Welcome User")

# function calling

greetuser()
greetuser()
greetuser()
```

```python
# function with parameter/argument

# what is parameter
# a variable defined in a function definition is known as parameter
# act as a placeholder

#function definition
def greetuser(name) :
  print("Welcome User :" , name)

# function calling
greetuser("John")
greetuser("Smith")
```

```python
def addition(a,b) :
  print(f"Addition of {a} and {b} is {a+b}")

addition(10,30)
addition(20,70)
```

```python
# Task
# using function predict the future career

import random

def predict_career(name) :
  professions = ["Doctor", "Pilot", "Police", "Engineer", "Housewife"]
  print(f"Hi user {name} your predicted profession is {random.choice(professions)}")

# function calling

predict_career("John")
predict_career("Smith")
```

```python
# def a function to check whether the entered no. is even or odd

def check_even_odd(num) :
  if num % 2 == 0:
    print("even")
  else :
    print("odd")

num = int(input("Enter a number : "))
check_even_odd(num)
```

```python
# function with parameter and with return statement
# a function with return statement is used to return a value

def checkEvenOdd(n) :
  return n%2==0

num = int(input("Enter a number : "))
res = checkEvenOdd(num)

if res:
  print("Even")
else :
    print("Odd")
```

```python
# check whether the string is palindrome or not
# madam  madam
# nitin  nitin
# yash   yash

def strReverse(text):
  return text [::-1]

a = "yash"
res = strReverse(a)

if a == res :
  print("Palindrome string")
else:
  print("Not a Palindrome String")
```
