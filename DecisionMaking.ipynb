```python
# Task
# Print The Largest

a = 19
b = 45
c = 27
d = 141
e = 67

largest = 0

if a>largest:          # 19>0(t)
  largest = a          # largest = 19
if b>largest:          # 45>19(t)
  largest = b          # largest = 45
if c>largest:          # 27>45(f)
  largest = c          # largest = 45
if d>largest:          # 45>141(f)
  largest = d          # largest = 141
if e>largest:          # 67>141(f)
  largest = e          # largest = 141


print(f"Largest is : {largest}")
```

```python
# Task
# Grading Allotment Application

# 100-85 : A grade
# 84-75 : B grade
# 74-60 : C grade
# 59-40 : D grade
# 39-0 : Fail

marks = int(input("Enter Marks: "))
if marks<=100 and marks>=85:
  print("A grade")
elif marks<84 and marks>=75:
  print("B grade")
elif marks<74 and marks>=60:
  print("C grade")
elif marks<59 and marks>=40:
  print("D grade")
elif marks<39 and marks>=0:
  print("Fail")
else:
  print("Invalid")
```

```python
# Chained Comparison
# help us to define a range
# Lower Limit <=  value   <=  Upper Limit


# here 85 <= marks <= 100 this expression means
#  a.) the value is less than or equal to 100    c1     is 93 <=100 (t)
#  b.) 85 means that I am less the define value  c2     is 85 is less than 93 (t)

print("Grading Allotment Application")
marks = int(input("Enter Your Marks :"))

if 85 <= marks <= 100:       #  83<=100 (t)   85 is it less than 83 (f)
   print("Grade A")
elif 75 <= marks <= 84:      # 83<=84  (t)    75 is it less than 83 (t)
  print("Grade B")
elif 60 <= marks <= 74:
  print("Grade C")
elif 40 <= marks <= 59:
  print("Grade D")
elif 0 <= marks <= 39:
  print("Fail")
else:
  print("Invalid")
```
